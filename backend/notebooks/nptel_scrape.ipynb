```python
data_jan22 = pd.read_csv("../data/raw/nptel-jan-apr-2022.csv")
data_jul22 = pd.read_csv("../data/raw/nptel-july-dec-2022.csv")
data_jan23 = pd.read_csv("../data/raw/nptel-jan-apr-2023.csv")
data_jul23 = pd.read_csv("../data/raw/nptel-july-dec-2023.csv")
data_jan24 = pd.read_csv("../data/raw/nptel-jan-apr-2024.csv")
data_jul24 = pd.read_csv("../data/raw/nptel-july-dec-2024.csv")
```

```python
req_cols = [
    "course id",
    "discipline",
    "course name",
    "sme name",
    "institute",  # Institute 1
    "\nco-ordinating institute",  # Institute 2
    "duration",
    "click here to join the course",  # URL
]


def preprocess(df: pd.DataFrame):
    data = df
    data.columns = map(str.lower, data.columns)
    # preprocessed data
    data = data[req_cols]
    data.rename(
        columns={
            "course id": "id",
            "institute": "institute 1",
            "\nco-ordinating institute": "institute 2",
            "click here to join the course": "url",
        },
        inplace=True,
    )
    return data


preprocessed_data = preprocess(data_jan22)
preprocessed_data
```

```python
st = set()
data_list = [
    data_jan22,
    data_jul22,
    data_jan23,
    data_jul23,
    data_jan24,
    data_jul24,
]

# required columns are present in the selected datasets
check = all(e in req_cols for e in data.columns for data in data_list)
check
```

```python
final_df = pd.DataFrame()
for data in data_list:
	preprocessed_data = preprocess(data)
	final_df = pd.concat([preprocessed_data])

final_df # contains data from 2022 - 2024
```
